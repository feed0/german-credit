```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import seaborn as sns

df = fetch_ucirepo(id=144).data.original
mem1 = df.memory_usage().sum()

# convert object types into categorical
df[df.select_dtypes('object').columns] = df.select_dtypes('object').astype('category')
mem2 = df.memory_usage().sum()

print(f"Memory reduction: {(1 - mem2/mem1) * 100:.2f}%")

sns.barplot({'before': mem1, 'after': mem2})
```
